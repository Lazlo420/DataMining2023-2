# 2017

## Cirugias

```python
cirugias=pd.read_csv('Datasets\\Cirugias\\ConsultaH_Intervenciones_Quirurgicas_2017_v1.csv',encoding='ISO-8859-1')
cirugias.replace('NE_0002', np.nan, inplace=True)
cirugias.replace('NE_0001', np.nan, inplace=True)
cirugias.fillna(0,inplace=True)
```

```python
cirugias=cirugias[cirugias['SECTOR']=='GOBIERNO REGIONAL']
cirugias.iloc[:,10:]=cirugias.iloc[:,10:].apply(pd.to_numeric,errors='coerce')
```

```python
cirugias.head()
```

```python
cirugias_to_sum = cirugias.columns[10:]
cirugias=cirugias.groupby(['CO_IPRESS','ANHO','MES'])[cirugias_to_sum].sum()
```

```python
cirugias=cirugias.reset_index()
```

## Procedimientos

```python
procedimientos=pd.read_csv("Datasets\\Procedimientos\ConsultaG_Procedimientos_2017_v1.csv",encoding='ISO-8859-1')
procedimientos.iloc[:,12]=procedimientos.iloc[:,12].apply(pd.to_numeric,errors='coerce')
procedimientos=procedimientos[procedimientos['SECTOR']=='GOBIERNO REGIONAL']
```

```python
procedimientos.head()
```

```python
procedimientos=procedimientos.groupby(['CO_IPRESS','MES','ANHO'])['TOTAL'].sum().reset_index()
```

## Complicaciones

```python
complicaciones=pd.read_csv("Datasets\\Complcaciones\\ConsultaF_Consolidado_Eventos_2017_v1.csv",encoding='ISO-8859-1')
complicaciones.replace('NE_0002', np.nan, inplace=True)
complicaciones.replace('NE_0001', np.nan, inplace=True)
complicaciones.fillna(0,inplace=True)
complicaciones=complicaciones[complicaciones['SECTOR']=='GOBIERNO REGIONAL']
```

```python
complicaciones.head()
```

```python
complicaciones.iloc[:,-1]=complicaciones.iloc[:,-1].apply(pd.to_numeric)
```

```python
complicaciones=complicaciones[complicaciones['EVENTOS']!=0]
```

```python
complicaciones = pd.pivot_table(complicaciones, values='CANT_EVENTOS', index=['ANHO', 'MES', 'UBIGEO', 'DEPARTAMENTO', 'PROVINCIA', 'DISTRITO', 'SECTOR', 'CATEGORIA', 'CO_IPRESS', 'RAZON_SOC'], columns='EVENTOS', aggfunc='sum', fill_value=0)
complicaciones.reset_index(inplace=True)
complicaciones.head()
```

```python
colums_to_sum=complicaciones.columns[-21:]
```

```python
complicaciones=complicaciones.groupby(['CO_IPRESS','MES','ANHO'])[colums_to_sum].sum().reset_index()
complicaciones.head()
```

## Nacimientos

```python
nacimientos=pd.read_csv("Datasets\\Nacimientos\ConsultaE_Consolidado_Partos_2017_v1_0.csv",encoding='ISO-8859-1')
nacimientos.replace('NE_0002', np.nan, inplace=True)
nacimientos.replace('NE_0001', np.nan, inplace=True)
nacimientos.fillna(0,inplace=True)
nacimientos=nacimientos[nacimientos['SECTOR']=='GOBIERNO REGIONAL']
```

```python
nacimientos.head()
```

```python
nacimientos.iloc[:,-3:]=nacimientos.iloc[:,-3:].apply(pd.to_numeric)
```

```python
nacimientos['Partos_CESAREA'] = 0
nacimientos['Partos_vivos_CESAREA'] = 0
nacimientos['Partos_muertos_CESAREA'] = 0
nacimientos['Partos_VAGINAL'] = 0
nacimientos['Partos_vivos_VAGINAL'] = 0
nacimientos['Partos_muertos_VAGINAL'] = 0

cesarea_condition = nacimientos['TIPO_PARTO'] == 'CESAREA'
vaginal_condition = nacimientos['TIPO_PARTO'] == 'VAGINAL'

nacimientos.loc[cesarea_condition, 'Partos_CESAREA'] = nacimientos.loc[cesarea_condition, 'PARTOS_TOTAL']
nacimientos.loc[vaginal_condition, 'Partos_VAGINAL'] = nacimientos.loc[vaginal_condition, 'PARTOS_TOTAL']

nacimientos.loc[cesarea_condition, 'Partos_vivos_CESAREA'] = nacimientos.loc[cesarea_condition, 'PARTOS_VIVOS']
nacimientos.loc[cesarea_condition, 'Partos_muertos_CESAREA'] = nacimientos.loc[cesarea_condition, 'PARTOS_MUERTOS']

nacimientos.loc[vaginal_condition, 'Partos_vivos_VAGINAL'] = nacimientos.loc[vaginal_condition, 'PARTOS_VIVOS']
nacimientos.loc[vaginal_condition, 'Partos_muertos_VAGINAL'] = nacimientos.loc[vaginal_condition, 'PARTOS_MUERTOS']

nacimientos.drop(['PARTOS_TOTAL','PARTOS_VIVOS','PARTOS_MUERTOS','ID_PARTO','TIPO_PARTO'],axis=1,inplace=True)
```

```python
nacimientos.head()
```

```python
nacimientos=nacimientos.groupby(['CO_IPRESS','MES','ANHO'])[['Partos_CESAREA','Partos_VAGINAL','Partos_vivos_CESAREA','Partos_vivos_VAGINAL','Partos_muertos_CESAREA','Partos_muertos_VAGINAL']].sum()
nacimientos=nacimientos.reset_index()
```

```python
nacimientos.head()
```

## HospxEsp

```python
espe=pd.read_csv("Datasets\\Espe\\ConsultaD1_Hospitalizaciones_Especialidad_2017_v1.csv",encoding='ISO-8859-1')
espe.replace('NE_0002', np.nan, inplace=True)
espe.replace('NE_0001', np.nan, inplace=True)
espe.fillna(0,inplace=True)
espe=espe[espe['SECTOR']=='GOBIERNO REGIONAL']
```

```python
espe.iloc[:,-7:]=espe.iloc[:,-7:].apply(pd.to_numeric)
espe=espe.groupby(['CO_IPRESS','MES','ANHO'])[['NRO_TOTAL_HOSPIT_ING','NRO_TOTAL_HOSPIT_EGR','NRO_TOTAL_ESTANCIAS','NRO_TOTAL_PACIENTES_CAMAS','NRO_TOTAL_CAMAS','NRO_TOTAL_CAMAS_DISPONIB','NRO_TOTAL_FALLECIDOS']].sum()
espe=espe.reset_index()
espe.head()
```

## Recursos

```python
recursos=pd.read_csv("Datasets\\Recursos\\ConsultaA_Recurso_Salud_2017_v1.csv",encoding='ISO-8859-1')
recursos.replace('NE_0002', np.nan, inplace=True)
recursos.replace('NE_0001', np.nan, inplace=True)
recursos.fillna(0,inplace=True)
recursos=recursos[recursos['SECTOR']=='GOBIERNO REGIONAL']
```

```python
recursos.iloc[:,10:26]=recursos.iloc[:,10:26].apply(pd.to_numeric)
recursos.head()
```

```python
columns_to_sum = recursos.columns[10:26]
recursos = recursos.groupby(['CO_IPRESS', 'MES','ANHO'])[columns_to_sum].sum()
recursos = recursos.reset_index()
recursos
```

## base final

```python
print(recursos.shape)
print(nacimientos.shape)
```

```python
df17 = pd.merge(recursos, nacimientos, on=['CO_IPRESS', 'MES', 'ANHO'], how='outer')
df17 = pd.merge(df17, cirugias, on=['CO_IPRESS', 'MES','ANHO'], how='outer')
df17 = pd.merge(df17, espe, on=['CO_IPRESS', 'MES','ANHO'], how='outer')
df17 = pd.merge(df17, procedimientos, on=['CO_IPRESS', 'MES','ANHO'], how='outer')
df17 = pd.merge(df17, complicaciones, on=['CO_IPRESS', 'MES','ANHO'], how='outer')
```
